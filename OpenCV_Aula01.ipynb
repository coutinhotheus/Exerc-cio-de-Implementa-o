```python
import cv2
#A função cv2.imshow é incompatível com Colab, por isso temos que usar cv2_imshow:
from google.colab.patches import cv2_imshow
```

```python
# Aqui decidimos não modificar a ordem dos pixels da imagem:
img = cv2.imread('imagem.jpg', cv2.IMREAD_UNCHANGED)
cv2_imshow(img)
```

```python
# Exemplo de uso de conversão de cores. Convertendo uma imagem de BGR para escala de cinza:
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
# Exemplo de uso de conversão de cores. Convertendo uma imagem de BGR para escala de RGB:
imgBGR = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2_imshow(imgBGR)
```

```python
# Função de desenho: desenhando círculo:
cv2.circle(img, (130, 130), 85, (255,0,255), 45 )
cv2_imshow(img)
```

```python
img = cv2.imread('imagem.jpg', cv2.IMREAD_UNCHANGED)
print(img.shape)

cv2.rectangle(img, (128, 128), (163, 172), (255, 255, 0, 0), 10)
cv2_imshow(img)

cv2.imwrite("imamge.jpg", img)
```

```python
# Desfocando (Blur) em uma imagem grayscale usando filtro Guassiano (padrão): 
imBlur = cv2.blur(gray,(45,45))
cv2_imshow(imBlur)
```

```python
from google.colab.patches import cv2_imshow
import cv2

cap = cv2.VideoCapture('video.mp4')
while cap.isOpened():
    ret, image = cap.read()

    if not ret:
      break

    cv2_imshow(image) # Note cv2_imshow em vez de cv2.imshow

    cv2.waitKey(1) & 0xff

cv2.destroyAllWindows()
cap.release()
```
